**About Dataset :-** This dataset is related to Big Basket online grocery products. The dataset is suitable for performing Exploratory Data Analysis (EDA) to understand pricing, discounts, ratings, and product distribution patterns.

**Measuring Discount on Product**

```python
df["discount_percent"] = (
    (df["market_price"] - df["sale_price"])
    / df["market_price"]) * 100
```

```python
df[["product", "market_price", "sale_price", "discount_percent"]].head()
```

```python
df["discount_percent"] = df["discount_percent"].round(2)
```

```python
df[["product", "discount_percent"]].head()
```

**Data Cleaning**

**Missing value**

```python
df.isnull().sum()
```

**Handling Missing Values**

**Filling Product Column with Mode**

```python
df["product"] = df["product"].fillna(df["product"].mode()[0])
```

**Filling Brand Column with Mode**

```python
df["brand"] = df["brand"].fillna(df["brand"].mode()[0])
```

**Filling Sale Price Column with Median**

```python
df["sale_price"] = df["sale_price"].fillna(df["sale_price"].median())
```

**Filling Rating Column with Median**

```python
df["rating"] = df["rating"].fillna(df["rating"].median())
```

**Filling Description Column with Mode**

```python
df["description"] = df["description"].fillna(df["description"].mode()[0])
```

**Filling Discount Percent Column with Median**

```python
df["discount_percent"] = df["discount_percent"].fillna(df["discount_percent"].median())
```

**Missing Values Check**

```python
df.isnull().sum()
```

**Outliers Detection Using IQR Method**

**Based on Sale Price**

```python
# **Sale Price**
Q1 = df["sale_price"].quantile(0.25)
Q3 = df["sale_price"].quantile(0.75)
# Calculate IQR
IQR = Q3 - Q1
# Calculate Limit
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

**Checking Outliers**

```python
outliers = df[
    (df["sale_price"] < lower_limit) |
    (df["sale_price"] > upper_limit)]
outliers
```

**Based on Market Price**

```python
# **Market Price**
Q1 = df["market_price"].quantile(0.25)
Q3 = df["market_price"].quantile(0.75)
# Calculate IQR
IQR = Q3 - Q1
# Calculate Limit
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python
# Checking Outliers
outliers = df[
    (df["market_price"] < lower_limit) |
    (df["market_price"] > upper_limit)]
outliers    
```

**Based on Rating**

```python
# **Rating**
Q1 = df["rating"].quantile(0.25)
Q3 = df["rating"].quantile(0.75)
# Calculate IQR
IQR = Q3 - Q1
# Calculate Limit
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python
# Checking Outliers
outliers = df[
    (df["rating"] < lower_limit) |
    (df["rating"] > upper_limit)]
outliers    
```

**Calculating Mean Values**

```python
sale_mean = df["sale_price"].mean()

market_mean = df["market_price"].mean()

rating_mean = df["rating"].mean()

print(sale_mean)

print(market_mean)

print(rating_mean)
```

**Replacing Sale Price Outliers with Mean**

```python
df["sale_price"] = np.where(
    (df["sale_price"] < lower_limit) |
    (df["sale_price"] > upper_limit),
    sale_mean,
    (df["sale_price"]))
```

```python
df["sale_price"]
```

**Replacing Market Price Outliers with Mean**

```python
df["market_price"] = np.where(
    (df["market_price"] < lower_limit) |
    (df["market_price"] > upper_limit),
    market_mean,
    (df["market_price"]))
```

```python
df["market_price"]
```

**Replacing Rating Outliers with Mean**

```python
df["rating"] = np.where(
    (df["rating"] < lower_limit) |
    (df["rating"] > upper_limit),
    rating_mean,
    (df["rating"]))
```

```python
df["rating"]
```
